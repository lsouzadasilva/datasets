### Lendo planilha com ("aba") específica

```python
import pandas as pd 

tabela = pd.read_excel("PLANILHOA.xlsx", sheet_name="BASE MATÉRIA PRIMA")

display(tabela)

# sheet_name = traz a aba do nosso dataframe
# usecols = e um selecionador de coluna 
# skiprows =  pula a linha desejada em nosso dataframe
```

### Tipo de filtros

```python
filter1 = tabela["Fornecedor"].isin(["CURTIDORA ITAUNA", "CURTUME DELLA TORRE"])

filter2 = (tabela["Fornecedor"] == "CURTIDORA ITAUNA") | (tabela["Fornecedor"] == "CURTUME DELLA TORRE")

# Esse meio de filtro ~ significa diferente e funcion melhor que !=
filtro = ~tabela["CÓDIGO"].isin(["CURTIDORA ITAUNA54508", "CURTUME DELLA TORRE34298"])
tabela = tabela[filtro]

display(tabela[filtro])
```

### Lendo planilha com ("aba") específica, com índice da aba

### Selecionando as colunas, pelo índice

```python
tabela = pd.read_excel("PLANILHOA.xlsx", sheet_name="BASE MATÉRIA PRIMA", usecols=[3, 4]) 
display(tabela)
```

### Buscando apenas a informação desejada 

```python
tabela = pd.read_excel("PLANILHOA.xlsx", sheet_name=None)
display(tabela["BASE MATÉRIA PRIMA"])
```

### Buscando todas as informações de um arquivo

```python
tabela = pd.read_excel("PLANILHOA.xlsx", sheet_name=None)
display(tabela)
```

### Selecionando as colunas, pelo "nome da coluna"

```python
tabela = pd.read_excel("PLANILHOA.xlsx", sheet_name="BASE MATÉRIA PRIMA", usecols="A:G")
display(tabela)
```

### Buscando informações de uma tabela fora de uma celula especifica

```python
tabela = pd.read_excel("PLANILHOA.xlsx", sheet_name="BASE MATÉRIA PRIMA", skiprows=7)
display(tabela)
```

```python
tabela = pd.read_excel("PLANILHOA.xlsx", sheet_name="BASE MATÉRIA PRIMA", usecols="A:G", skiprows=8)
display(tabela)
```

### Selecionando quantas linhas vc quer ver no dataframe

```python
tabela = pd.read_excel("PLANILHOA.xlsx", sheet_name="BASE MATÉRIA PRIMA", nrows=7)
display(tabela)
```
